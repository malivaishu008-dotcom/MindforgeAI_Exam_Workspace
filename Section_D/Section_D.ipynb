# Data Analytics Interpretation

### 1. Strongest Department
The strongest department is **Computer Engineering** because it has higher readiness scores, better attendance, and stronger assignment/lab performance compared to other departments.

### 2. Students Needing Academic Support
Students needing support are those with low attendance, fewer task completions, and low assignment/lab scores.  
Example: **Harshada** needs support because of low attendance (43%), fewer tasks completed (2), and low scores.

### 3. Dataset Limitation
The dataset is small and contains only 8 students, so it may not represent the performance of all students.

### 4. Extra Useful Columns
Additional columns that can improve analysis:
- Previous semester marks
- Attendance history
- Internship/project experience
- Skill assessment scores
- Feedback from mentors
