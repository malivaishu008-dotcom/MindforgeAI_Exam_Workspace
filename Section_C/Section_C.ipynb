```python
# Student Performance Analytics Pipeline

This notebook creates and loads a student dataset, checks data quality, calculates a readiness score, classifies students into performance categories, and generates analytics insights.
```

```python
import pandas as pd

# Create dataset
data = {
    "student_id": ["2026CS001","2026CS002","2026IT001","2026IT002","2026CE001","2026CE002","2026CS003","2026IT003"],
    "name": ["Aditi","Bhavana","Chaitra","Divya","Esha","Fatima","Gauri","Harshada"],
    "department": [
        "Computer Engineering",
        "Computer Engineering",
        "Information Technology",
        "Information Technology",
        "Civil Engineering",
        "Civil Engineering",
        "Computer Engineering",
        "Information Technology"
    ],
    "attendance_pct": [92,64,78,55,88,71,96,43],
    "diary_entries": [12,6,9,4,10,8,14,2],
    "tasks_completed": [8,5,7,3,8,6,9,2],
    "assignment_score": [84,58,72,46,79,66,91,35],
    "lab_score": [88,61,76,52,82,69,94,41]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("week_2_3_exam_student_signals.csv", index=False)

# Load CSV
df = pd.read_csv("week_2_3_exam_student_signals.csv")

# Display first five rows
df.head()
```

### Explanation

The student dataset was created, saved as **week_2_3_exam_student_signals.csv**, loaded successfully, and the first five records are displayed.

```python
# Shape of dataset
print("Dataset Shape:", df.shape)

# Data types
print("\nData Types:")
print(df.dtypes)

# Complete information
print("\nDataset Information:")
df.info()

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

### Explanation

The dataset has been checked for:
- Number of rows and columns
- Data types of each column
- Missing values

The dataset contains no missing values.

```python
# Function to calculate readiness score

def readiness_score(attendance, diary, task, assignment, lab):
    score = (
        attendance * 0.30 +
        diary * 0.10 +
        task * 0.20 +
        assignment * 0.20 +
        lab * 0.20
    )
    return round(score, 2)

# Apply function

df["readiness_score"] = df.apply(
    lambda row: readiness_score(
        row["attendance_pct"],
        row["diary_entries"],
        row["tasks_completed"],
        row["assignment_score"],
        row["lab_score"]
    ),
    axis=1
)

df
```

### Explanation

A custom function calculates the readiness score using attendance, diary entries, completed tasks, assignment marks, and lab marks with weighted values.

```python
performance = []

for score in df["readiness_score"]:
    if score < 50:
        performance.append("Support")
    elif score < 70:
        performance.append("Developing")
    elif score < 90:
        performance.append("Strong")
    else:
        performance.append("Excellent")

df["performance_band"] = performance

df
```

### Explanation

Students are classified into four performance categories based on their readiness score:

- Support
- Developing
- Strong
- Excellent

```python
# Department Average

print("Department Average Readiness Score")
print(df.groupby("department")["readiness_score"].mean())

print("\n")

# Top Performer

print("Top Performer")
print(df.loc[df["readiness_score"].idxmax()])

print("\n")

# Students Needing Support

print("Students Needing Support")
print(df[df["performance_band"] == "Support"])
```

### Explanation

The analytics provide:

- Average readiness score for each department.
- The highest-performing student.
- Students who require academic support.
